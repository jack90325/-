```python
#importmodules
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
#%matplotlibinline
import seaborn as sns 

#import dataset
df_train = pd.read_csv("train.csv")

#observing dataset 
df_train.head()
df_train.describe().T
```

將對訓練完全沒有幫助的ID跟名字都去除

```python
df_train.drop('PassengerId', axis=1, inplace=True)
df_train.drop('Name', axis=1, inplace=True)
df_train.head()
```

```python
df_train.isnull().sum()
```

將無法填補的資料整行去除

```python
df_train.dropna(axis=0,subset = ['HomePlanet'], inplace=True)
df_train.dropna(axis=0,subset = ['CryoSleep'], inplace=True)
df_train.dropna(axis=0,subset = ['Cabin'], inplace=True)
df_train.dropna(axis=0,subset = ['Destination'], inplace=True)
df_train.dropna(axis=0,subset = ['VIP'], inplace=True)
df_train.head()
```

將消費的空值都以該列的平均值進行填補

```python
df_train['RoomService'].fillna(value=df_train['RoomService'].mean(), inplace=True)
df_train['FoodCourt'].fillna(value=df_train['FoodCourt'].mean(), inplace=True)
df_train['ShoppingMall'].fillna(value=df_train['ShoppingMall'].mean(), inplace=True)
df_train['Spa'].fillna(value=df_train['Spa'].mean(), inplace=True)
df_train['VRDeck'].fillna(value=df_train['VRDeck'].mean(), inplace=True)
df_train.isnull().sum()
```

將非數值屬性的表現方式用get_dummies()改成0或1

```python
df_train=pd.get_dummies(data=df_train,columns=['VIP','HomePlanet','Destination','CryoSleep','Transported'])
df_train.head()
df_train.drop('VIP_False',axis=1,inplace=True)
df_train.drop('CryoSleep_False',axis=1,inplace=True)
df_train.drop('Transported_False',axis=1,inplace=True)
df_train.head()
```

```python
df_train.isnull().sum()
```

將Age的空值用中位數來填補

```python
median = df_train['Age'].median()
df_train['Age'].fillna(median, inplace=True)
df_train.head()
```

```python
df_train.isnull().sum()
```

將Cabin分成兩個欄位，一個是坐第幾排，一個是是否靠走道

```python
df_train['side_or_not'] = 0
for i in range(len(df_train)):
    SoP = df_train.iloc[i,0]
    df_train.iloc[i,16] = SoP[-1]
    s = df_train.iloc[i,0]
    df_train.iloc[i,0] = s[0]
df_train.head()
```

```python
df_train=pd.get_dummies(data=df_train,columns=['Cabin','side_or_not'])
df_train.drop('side_or_not_P',axis=1,inplace=True)
df_train.head()
```

最後將資料分割進行訓練，用LogisticRegression來訓練模型

```python
X=df_train.drop(['Transported_True'],axis=1)
y=df_train['Transported_True']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=24)
```

```python
from sklearn.linear_model import LogisticRegression 
lr = LogisticRegression()
lr.fit(X_train,y_train)
predictions = lr.predict(X_test)
```

輸出訓練模型

```python
import joblib 
joblib.dump(lr,'Titanic-LR-20230415.pkl',compress=3)
```
